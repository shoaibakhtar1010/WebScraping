# combining data sources

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def get_financial_info(symbol):
    from bs4 import BeautifulSoup
    url="https://finance.yahoo.com/quote/AAPL?p=AAPL"
    response=requests.get(url)
    t=response.text
    soup=BeautifulSoup(t,features="html.parser")
    finalName="1y Target Est"
    trs= soup.find_all("tr")
    names=[]
    values=[]
    for i in range(len(trs)):
        for j in range(len(trs[i])):
            if j==0: #name
                try:
                    name=trs[i].contents[j].text
                    names.append(name)
                except:
                    continue
            if j==1: # value
                try:
                    value=trs[i].contents[j].text
                    values.append(value)
                except:
                    continue
    
        if name==finalName: # break the loop and doesnot move towards further tr tags
            break
    return names,values
```

```python
symbol="ALLE"
names,values=get_financial_info(symbol)
for name,value in zip(names,values):
    print(name,value)
```

```python
def get_company_list():
    url="https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"
    r=requests.get(url)
    text=r.text
    soup=BeautifulSoup(text,features="html.parser")
    tbody=soup.find_all("tbody")
    endsymbol="ZTS"
    symbols=[]
    for i in range(2,len(tbody[0].contents),2):
        symbol= tbody[0].contents[i].contents[1].text.strip("\n")
        symbols.append(symbol)
        if symbol==endsymbol:
            break
    return symbols
```

```python
ls=get_company_list()
ls
```

```python
def main():
    data={"symbol":[],
          "metric":[],
          "value":[]
          }
    company_symbol_list=get_company_list()
    for symbol in company_symbol_list:
        names,values=get_financial_info(symbol)
        for i in range(len(names)):
            data["symbol"].append(symbol)
            data["metric"].append(names[i])
            data["value"].append(values[i])
    df=pd.DataFrame(data)
    df.to_csv("3_combining.csv")
    return df
```

```python
df=main()
df
```
